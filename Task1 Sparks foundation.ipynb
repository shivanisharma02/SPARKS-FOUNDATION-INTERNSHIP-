## *Task: Predicting the percentage of the student based on the number of study hours* 

### *Dataset used: http://bit.ly/w-data*

#### **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

#### **Loading the Dataset**

```python
df = pd.read_csv('http://bit.ly/w-data')
df.head()
```

#### **Obtaining the dimensions of the dataset**

```python
df.shape
```

```python
df.describe()
```

```python
# Checking for null values in the datset
df.isnull().sum()
```

##### *Turns out there are no nul values in our datset*

### **DATA VISUALIATION**

```python
#Plotting the data
plt.figure()
plt.plot(df['Hours'].values, df['Scores'].values, 'o', color='green')
plt.grid(True)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Study Hours vs Scores')
plt.show()
```

#### *As the data can be seen to be highly correlated hence we can make a linear regression model.*

```python
#Dividing the dataset into independant and dependant variables
X = df.iloc[:, :1].values  #dependant variable
y = df.iloc[:,-1:].values  #independant variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

### **Creating a Linear Regression Model**

```python
model = LinearRegression()
```

### **Training The model**

```python
model.fit(X_train, y_train)  
line = model.coef_*X+model.intercept_  
```

### **Plotting the Data**

```python
plt.scatter(X, y, color = 'green')
plt.plot(X, line, color = 'red')
plt.grid(True)
plt.xlabel('Hours')
plt.ylabel('Score')
plt.title('Study Hours vs Scores')
plt.show()
```

### **PREDICTION**

```python
regr_predict = model.predict(X_test)
```

#### *Predicting the Scores for someone who studies for 9.25 hours*

```python
hours =  [[9.25]]
result = model.predict(hours)
print("No. of Hours = {}".format(hours))
print("Score Predicted = {}".format(result[0]))
```

#### *The predicted score is 95.228*

### **PERFORMANCE EVALUATION**

```python
print('Error:', metrics.mean_absolute_error(y_test, regr_predict))
```

#### *The Mean Absolute Error of the model turns out to be 7.88*
